```python
# IMPORT LIBRARIES

import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
```

```python
# READ DATA

Data = pd.read_csv('movies.csv')
```

### DATA CLEANING

```python
# QUERY COLUMN DATA

Data.query('budget=="NAN"')
```

```python
# MISSING VALUES

Data.isna().sum()
```

```python
# DROP ALL MISSING DATA

Data = Data.dropna()
```

```python
# REVIEW MISSING VALUES

Data.isna().sum()
```

```python
# COLUMNS CONVERT TO INTEGER

Data['budget'] = Data['budget'].astype('int64')
Data['gross'] = Data['gross'].astype('int64')

Data.dtypes
```

```python
# VIEW DATA AGAIN

Data
```

### DATA EXPLORATION

```python
# HIGHEST GROSSING MOVIE

Data.sort_values(by = ['gross'], ascending = False)
```

```python
# CHECK DUPLICATE

Data.loc[Data.duplicated(subset = ['name', 'year', 'budget'])]
```

```python
# QUERY COLUMN DATA

Data.query('name == "Impulse"')
```

```python
# SUPOSSITIONS
#1 BUDGET (HIGH CORRELATION WITH GROSS EARNINGS)
#2 DIRECTOR 
#3 GENRE
```

### DATA VISUALIZATION

```python
# SCATTER PLOT

ax = Data.plot(kind = 'scatter', x = 'gross', y = 'budget', figsize = (12,8))
ax.set_xlabel('Gross earnings')
```

```python
# REG PLOT

plt.figure(figsize=(12, 8))
sns.regplot(data = Data, x = 'budget', y = 'gross', scatter_kws = {'color':'red'}, line_kws = {'color':'black'})
```

```python
# CORRELATION

Data.dropna()
Data.corr()
```

```python
# ASIGN A NEW VARIABLE 

Data_corr = Data.corr()
```

```python
# SUPPOSITIONS
#1 BUDGET TO A CERTAIN HAS A HIGH DEGREE OF INFLUENCE OR CORRELATES MOST WITH PROFIT GROWTH

Data_corr.sort_values(by = ['budget'], ascending = False)
```

```python
# RENAME COLUMN

Data_corr = Data_corr.rename(columns = {'gross': 'earnings'})
Data_corr
```

```python
# HEAT MAP - (CORRELATION DEFAULT)

sns.heatmap(Data_corr, annot = True)
plt.title('Movie Data Corr')
plt.xlabel('Attribute')
```

```python
# HEAT MAP - (CORRELATION KENDALL)

Data_corr = Data.corr(method = 'kendall')
sns.heatmap(Data_corr, annot = True)
plt.title('Movie Data Corr')
plt.xlabel('Attribute')
```

```python
# CONVERT OBJECT DATA IN NUMBER DATA TO REANALYZE

Data_numerized = Data.copy()

for col_names in Data_numerized.columns:
    if(Data_numerized[col_names].dtype == 'object'):
        Data_numerized[col_names] = Data_numerized[col_names].astype('category')
        Data_numerized[col_names] = Data_numerized[col_names].cat.codes
        
Data_numerized
```

```python
Data
```

```python
# CORRELATION BETWEEN NUMBERS

Data_numerized_corr = Data_numerized.corr()

Data_numerized_corr
```

```python
# SORT BY GROSS

Data_numerized_corr.sort_values(by = ['gross'], ascending = False)
```

```python
# HEAT MAP (DATA NUMERIZED)- (CORRELATION DEFAULT)

plt.figure(figsize = (12,8))
sns.heatmap(data = Data_numerized_corr)
plt.title('Movie Data Corr')
plt.xlabel('Attribute')
```

```python
# HEAT MAP (DATA NUMERIZED)- (CORRELATION DEFAULT)

plt.figure(figsize = (12,8))
sns.heatmap(Data_numerized_corr, annot = True)
plt.title('Movie Data Corr')
plt.xlabel('Attribute')
```

```python
# CORRELATION DATA EVEN FURTHER

Data_numerized_corr_unstack = Data_numerized_corr.unstack()

Data_numerized_corr_unstack_pair = Data_numerized_corr_unstack.sort_values()

Data_numerized_corr_unstack_pair
```
